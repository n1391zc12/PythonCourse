```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats

df = pd.DataFrame({'Height':[185.0, 174.2, 178.4, 179.1, 180.0, 177.5, 176.4, 190.3, 177.2, 176.7,
                             188.7, 177.2, 180.1, 188.6, 165.8, 183.0, 177.5, 181.2, 174.5, 171.6],
                   'Weight':[80.3, 90.7, 84.3 ,77.2, 75.32, 77.4, 80.2, 86.7, 76.8, 74.6,
                             90.4, 77.5, 77.8, 79.4, 69.3, 77.9, 80.6, 67.2, 73.1, 79.4],
                   'BMI':[23.46, 29.88, 26.48, 24.06, 23.24, 24.56, 25.77, 23.94, 24.45, 23.89,
                          25.38, 24.68, 23.98, 22.32, 25.20, 23.26, 25.58, 20.46, 24.00, 26.96]})
print ('Min: ' + str(df['BMI'].min()))
print ('Mode: ' + str(df['BMI'].mode()[0]))
print ('Median: ' + str(df['BMI'].median()))
print ('Mean: ' + str(df['BMI'].mean()))
print ('Max: ' + str(df['BMI'].max()))

numcols = ['Height', 'Weight', 'BMI']
for col in numcols:
    print(df[col].name + ' range: ' + str(df[col].max() - df[col].min()))

bmi = df['BMI']
density = stats.gaussian_kde(bmi)
n, x, _ = plt.hist(bmi, histtype='step', normed=True, bins=20)  
plt.plot(x, density(x)*5)
plt.axvline(bmi.mean(), color='magenta', linestyle='dashed', linewidth=2)
plt.axvline(bmi.median(), color='green', linestyle='dashed', linewidth=2)

plt.show()
```

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats

df = pd.DataFrame(np.random.randn(100000, 1), columns=['BMI'])

# Plot the distribution as a histogram with a density curve
bmi = df['BMI']
density = stats.gaussian_kde(bmi)
n, x, _ = plt.hist(bmi, color='lightgrey', normed=True, bins=100)  
plt.plot(x, density(x))

# Get the mean and standard deviation
s = df['BMI'].std()
m = df['BMI'].mean()

# Annotate 1 stdev
x1 = [m-s, m+s]
y1 = [0.25, 0.25]
plt.plot(x1,y1, color='magenta')
plt.annotate('1s (68.26%)', (x1[1],y1[1]))

# Annotate 2 stdevs
x2 = [m-(s*2), m+(s*2)]
y2 = [0.05, 0.05]
plt.plot(x2,y2, color='green')
plt.annotate('2s (95.45%)', (x2[1],y2[1]))

# Annotate 3 stdevs
x3 = [m-(s*3), m+(s*3)]
y3 = [0.005, 0.005]
plt.plot(x3,y3, color='orange')
plt.annotate('3s (99.73%)', (x3[1],y3[1]))

# Show the location of the mean
plt.axvline(bmi.mean(), color='grey', linestyle='dashed', linewidth=1)

plt.show()
```

```python
%matplotlib inline
import pandas as pd
from matplotlib import pyplot as plt

df = pd.DataFrame({'Height':[185.0, 174.2, 178.4, 179.1, 180.0, 177.5, 176.4, 190.3, 177.2, 176.7,
                             188.7, 177.2, 180.1, 188.6, 165.8, 183.0, 177.5, 181.2, 174.5, 171.6],
                   'Weight':[80.3, 90.7, 84.3 ,77.2, 75.32, 77.4, 80.2, 86.7, 76.8, 74.6,
                             90.4, 77.5, 77.8, 79.4, 69.3, 77.9, 80.6, 67.2, 73.1, 79.4],
                   'BMI':[23.46, 29.88, 26.48, 24.06, 23.24, 24.56, 25.77, 23.94, 24.45, 23.89,
                          25.38, 24.68, 23.98, 22.32, 25.20, 23.26, 25.58, 20.46, 24.00, 26.96]})

# Create a scatter plot of Salary vs Grade
df.plot(kind='scatter', title='Height vs Weight', x='Weight', y='BMI')
plt.plot(np.unique(df['Weight']), np.poly1d(np.polyfit(df['Weight'], df['BMI'], 1))(np.unique(df['Weight'])))
plt.show()
```
